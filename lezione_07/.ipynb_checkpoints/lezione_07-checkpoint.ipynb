## <span style="color:limegreen"> Lezione 7</span>

In questa esercitazione verranno utilizzati i dati prodotti dal codice Monte Carlo NVT per studiare il tempo di autocorrelazione e l'incertezza al variare della dimensione dei blocchi dell'energia interna e della pressione di un sistema nelle seguenti condizioni termodinamiche:

- fase solida: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
- fase liquida: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
- fase gassosa: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

Al codice è stato aggiunto anche il calcolo della $g(r)$, che può essere confrontata con quella prodotta nel codice di dinamica molecolare.

```python
import matplotlib                                        
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
def exponential(x, a, b, c):
    return a * np.exp(-b * x) + c

tmax = 100000                       #tmax = 10**5 (dovrebbe...)
tcut = 1000                        #parametri per calcolo autocorrelazione

L = []                             #dimensione blocchi per studio data blocking
for i in range(1,501):                              
    L.append(10*(i))
```

## <span style="color:red"> Stato: Solido </span>

### Energia Interna Istantanea

```python
fig=plt.figure(figsize=(15, 5))                     #sx: larghezza; dx: Lunghezza

ene_solid = np.loadtxt(r"Data/solid/instant_energy.solid", unpack='true')       #attenzione: non ci sono spazi nei files!
plt.plot(ene_solid, color='blue')
plt.grid(True)
plt.xlabel('N step')
plt.ylabel("U/N")
plt.title('Energia interna istantanea (solido)')

plt.show()
```

#### Autocorrelazione

```python
sum4 = 0.0
sum5 = 0.0
for t1 in range(0, tmax):
    sum4 += (ene_solid[t1])*(ene_solid[t1])
    sum5 += ene_solid[t1]
    
den = ((1./tmax)*(sum4 - (1./tmax)*sum5*sum5)) #denominatore nella formula dell'autocorrelazione

def X_en_solid(t):        
    sum1 = 0.0
    sum2 = 0.0
    sum3 = 0.0      # in ordine come compaiono nella formula, prima num poi den
    
    for t1 in range(0, tmax - t):
        sum1 += ene_solid[t1]*ene_solid[t1+t]
        sum2 += ene_solid[t1]
        sum3 += ene_solid[t1+t]                  #ATTENZIONE, si può ottimizzare molto!!!!!
                                                    
    return ((1./(tmax-t))*(sum1 - (1./(tmax-t))*sum2*sum3)/den)    
            
T = []
X_en_vett_s = []

for i in range(0, tcut):
    T.append(i)
    X_en_vett_s.append(X_en_solid(T[i]))
    
def exponential(x, a, b, c):
    return a * np.exp(-b * x) + c
```

```python
fig=plt.figure(figsize=(15, 5))                     #sx: larghezza; dx: Lunghezza

plt.plot(T, X_en_vett_s, color='c')
popt, pcov = curve_fit(exponential, T, X_en_vett_s)
y_fit = []

for i in range(0, tcut):
    y_fit.append(exponential(i,popt[0],popt[1],popt[2]))

plt.plot(T,y_fit, color='r') # plotting fitted function 
print("optimized parameters [a,b,c] =")
print(popt)
print("parameters uncertainty =")
print(np.sqrt(np.diagonal(pcov)))
plt.xlabel('T')
plt.ylabel("X")
plt.title("Autocorrelazione (energia - solido)")

print('\ntempo di autocorrelazione =')
print(1./popt[1])

plt.grid(True)
plt.show()
```

#### Errore Data Blocking

```python
err_en_s = []
for l in L:
    N_blocchi = int(10**5/l)                  #prendo intero più vicino...
    media = 0.0
    media2 = 0.0
    for i in range(0,N_blocchi):                     #scorrono i blocchi
        sum1 = 0.0
        for n in range(0,l):          #eseguo il blocco
            sum1 += ene_solid[n + l*i]
        sum1 = sum1*(1./l)
        sum2 = sum1*sum1    #quadrato della media in ogni blocco
        media = ((media*i)+sum1)*(1./(i+1))
        media2 = ((media2*i)+sum2)*(1./(i+1))
    
    err_en_s.append( np.sqrt((media2-media*media)/(N_blocchi)) ) #salva solo ultima incertezza allo scorrere dei blocchi
```

```python
fig=plt.figure(figsize=(15,5))                     #sx: larghezza; dx: Lunghezza

plt.plot(L, err_en_s, color='lime')
plt.grid(True)
plt.xlabel('N blocchi')    
plt.ylabel('incertezza')
plt.title('Errore statistico in funzione del numero di blocchi (energia - solido)')
plt.show()
```

```python
err_en_g = []
for l in L:
    N_blocchi = int(10**5/l)                  #prendo intero più vicino...
    media = 0.0
    media2 = 0.0
    for i in range(0,N_blocchi):                     #scorrono i blocchi
        sum1 = 0.0
        for n in range(0,l):          #eseguo il blocco
            sum1 += ene_gas[n + l*i]
        sum1 = sum1*(1./l)
        sum2 = sum1*sum1    #quadrato della media in ogni blocco
        media = ((media*i)+sum1)*(1./(i+1))
        media2 = ((media2*i)+sum2)*(1./(i+1))
    
    err_en_g.append( np.sqrt((media2-media*media)/(N_blocchi)) ) #salva solo ultima incertezza allo scorrere dei blocchi
```

```python
fig=plt.figure(figsize=(15,5))                     #sx: larghezza; dx: Lunghezza

plt.plot(L, err_en_g, color='lime')
plt.grid(True)
plt.xlabel('N_blocchi')    
plt.ylabel('incertezza')
plt.title('Errore statistico in funzione del numero di blocchi')
plt.show()
```

#### Autocorrelazione

```python
sum4 = 0.0
sum5 = 0.0
for t1 in range(0, tmax):
    sum4 += (pres_gas[t1])*(pres_gas[t1])
    sum5 += pres_gas[t1]
    
den = ((1./tmax)*(sum4 - (1./tmax)*sum5*sum5)) #denominatore nella formula dell'autocorrelazione

def X_pres_gas(t):        
    sum1 = 0.0
    sum2 = 0.0
    sum3 = 0.0      # in ordine come compaiono nella formula, prima num poi den
    
    for t1 in range(0, tmax - t):
        sum1 += pres_gas[t1]*pres_gas[t1+t]
        sum2 += pres_gas[t1]
        sum3 += pres_gas[t1+t]                  #ATTENZIONE, si può ottimizzare molto!!!!!
                                                    
    return ((1./(tmax-t))*(sum1 - (1./(tmax-t))*sum2*sum3)/den)    
            
T = []
X_pres_vett_g = []

for i in range(0, tcut):
    T.append(i)
    X_pres_vett_g.append(X_pres_gas(T[i]))
```

```python
fig=plt.figure(figsize=(15,5))                     #sx: larghezza; dx: Lunghezza
plt.plot(T, X_pres_vett_g, color='c')
popt, pcov = curve_fit(exponential, T, X_pres_vett_g)
x = np.linspace(0,tcut,endpoint=True)
y_fit = exponential(x,popt[0],popt[1],popt[2])
plt.plot(x,y_fit, color='r') # plotting fitted function 
print("optimized parameters [a,b,c] =")
print(popt)
print("parameters uncertainty =")
print(np.sqrt(np.diagonal(pcov)))
plt.xlabel('T')
plt.ylabel("X")
plt.title("Autocorrelazione (pressione - gas)")

print('tempo di autocorrelazione =')
print(1./popt[1])

plt.show()
```
