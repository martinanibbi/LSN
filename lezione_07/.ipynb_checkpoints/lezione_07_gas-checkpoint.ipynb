```python
import matplotlib                                        
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

## <span style="color:red"> Stato: gas </span>

### Energia Interna Istantanea

```python
fig=plt.figure(figsize=(15, 5))                     #sx: larghezza; dx: Lunghezza

ene_gas = np.loadtxt(r"Data/gas/instant_energy.gas", unpack='true')       #attenzione: non ci sono spazi nei files!
plt.plot(ene_gas, color='blue')
plt.grid(True)
plt.xlabel('N step')
plt.ylabel("U/N")
plt.title('Energia interna istantanea (gas)')

plt.show()
```

#### Autocorrelazione

```python
tmax = 10000                       #tmax = 10**5 (dovrebbe...)
tcut = 9999
sum4 = 0.0
sum5 = 0.0
for t1 in range(0, tmax):
    sum4 += (ene_gas[t1])*(ene_gas[t1])
    sum5 += ene_gas[t1]
    
den = ((1./tmax)*(sum4 - (1./tmax)*sum5*sum5)) #denominatore nella formula dell'autocorrelazione

def X_en_gas(t):        
    sum1 = 0.0
    sum2 = 0.0
    sum3 = 0.0      # in ordine come compaiono nella formula, prima num poi den
    
    for t1 in range(0, tmax - t):
        sum1 += ene_gas[t1]*ene_gas[t1+t]
        sum2 += ene_gas[t1]
        sum3 += ene_gas[t1+t]                  #ATTENZIONE, si può ottimizzare molto!!!!!
                                                    
    return ((1./(tmax-t))*(sum1 - (1./(tmax-t))*sum2*sum3)/den)    
            
fig=plt.figure(figsize=(16, 7))                     #sx: larghezza; dx: Lunghezza

T = []
X_en_vett = []
y_fit = []

for i in range(0, tcut):
    T.append(i)
    X_en_vett.append(X_en_gas(T[i]))
    
def exponential(x, a, b, c):
    return a * np.exp(-b * x) + c

plt.plot(T, X_en_vett, color='c')
popt, pcov = curve_fit(exponential, T, X_en_vett)

for i in range(0, tcut):
    y_fit.append(exponential(i,popt[0],popt[1],popt[2]))
    
plt.plot(T,y_fit, color='r') # plotting fitted function 
print("optimized parameters [a,b,c] =")
print(popt)
print("parameters uncertainty =")
print(np.sqrt(np.diagonal(pcov)))
plt.xlabel('T')
plt.ylabel("X")
plt.title("Autocorrelazione (gas)")

plt.show()
```

#### Errore Data Blocking

```python
L = []
for i in range(1,501):                              
    L.append(10*(i))

fig=plt.figure(figsize=(15, 5))                     #sx: larghezza; dx: Lunghezza
errore = []
for l in L:
    N_blocchi = int(10**5/l)                  #prendo intero più vicino...
    media = 0.0
    media2 = 0.0
    for i in range(0,N_blocchi):                     #scorrono i blocchi
        sum1 = 0.0
        for n in range(0,l):          #eseguo il blocco
            sum1 += ene_gas[n + i*l]
        sum1 = sum1*(1./l)
        sum2 = sum1*sum1    #quadrato della media in ogni blocco
        media = ((media*i)+sum1)*(1./(i+1))
        media2 = ((media2*i)+sum2)*(1./(i+1))
    
    errore.append( np.sqrt((media2-media*media)/(N_blocchi)) ) #salva solo ultima incertezza allo scorrere dei blocchi

plt.plot(L, errore, color='lime')
plt.grid(True)
plt.xlabel('N_blocchi')    
plt.ylabel('incertezza')
plt.title('Errore statistico in funzione del numero di blocchi')
plt.show()
```

### Pressione Istantanea

```python
fig=plt.figure(figsize=(15, 5))                     #sx: larghezza; dx: Lunghezza

pres_gas = np.loadtxt(r"Data/gas/instant_pressure.gas", unpack='true')       #attenzione: non ci sono spazi nei files!
plt.plot(pres_gas, color='blue')
plt.grid(True)
plt.xlabel('N step')
plt.ylabel("P")
plt.title('pressione istantanea (gas)')

plt.show()
```

#### Autocorrelazione

```python
tmax = 10000                       #tmax = 10**5 (dovrebbe...)
tcut = 9999
sum4 = 0.0
sum5 = 0.0
for t1 in range(0, tmax):
    sum4 += (pres_gas[t1])*(pres_gas[t1])
    sum5 += pres_gas[t1]
    
den = ((1./tmax)*(sum4 - (1./tmax)*sum5*sum5)) #denominatore nella formula dell'autocorrelazione

def X_pres_gas(t):        
    sum1 = 0.0
    sum2 = 0.0
    sum3 = 0.0      # in ordine come compaiono nella formula, prima num poi den
    
    for t1 in range(0, tmax - t):
        sum1 += pres_gas[t1]*pres_gas[t1+t]
        sum2 += pres_gas[t1]
        sum3 += pres_gas[t1+t]                  #ATTENZIONE, si può ottimizzare molto!!!!!
                                                    
    return ((1./(tmax-t))*(sum1 - (1./(tmax-t))*sum2*sum3)/den)    
            
fig=plt.figure(figsize=(16, 7))                     #sx: larghezza; dx: Lunghezza

T = []
X_pres_vett = []

for i in range(0, tcut):
    T.append(i)
    X_pres_vett.append(X_pres_gas(T[i]))
    
def exponential(x, a, b, c):
    return a * np.exp(-b * x) + c

plt.plot(T, X_pres_vett, color='c')
popt, pcov = curve_fit(exponential, T, X_pres_vett)
x = np.linspace(0,tcut,endpoint=True)
y_fit = exponential(x,popt[0],popt[1],popt[2])
plt.plot(x,y_fit, color='r') # plotting fitted function 
print("optimized parameters [a,b,c] =")
print(popt)
print("parameters uncertainty =")
print(np.sqrt(np.diagonal(pcov)))
plt.xlabel('T')
plt.ylabel("X")
plt.title("Autocorrelazione (gas)")

plt.show()
```
