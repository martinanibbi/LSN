### Esercizio 02.1
- calcolare l'integrale $$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$ con il metodo del Data Blocking.

Il valor medio dell'integrale è stato calcolato attraverso un campionamento uniforme e un *Importance Sampling* con la distribuzione normalizzata: $$p(x) = 2(1-x)$$
$p(x)$ è l'approssimazione lineare dell'integranda nell'intervallo $[0,1]$ e dunque, essendo più simile alla funzione da integrare rispetto alla distribuzione uniforme, con questo metodo si raggiunge una maggior precisione.

Con il metodo della funzione inversa si ottiene che per campionare $p(x)$ è necessario estrarre $z$ uniformemente ed imporre: $$x = 1 - \sqrt{1-z}$$

Infine, la funzione $f(x)$ si ottiene dividendo l'integranda originale e la distribuzione $p(x)$.

L'errore è stato calcolato con entrambi i campionamenti attraverso la deviazione standard della media.

Numero totale di numeri estratti: $M=10^5$. Blocchi: $N=10^2$

### Distribuzione uniforme

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
fig=plt.figure(figsize=(15,5))
N_blocchi, I, error = np.loadtxt(r"Data/integrale_uniforme.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(N_blocchi,I,yerr=error, color='limegreen')
plt.xlabel('N blocchi')
plt.ylabel('Integrale')
plt.title('I con distribuzione uniforme')
plt.axhline(y=1,color='forestgreen')
plt.grid(True)
plt.show()
```

### Importance Sampling

```python
fig=plt.figure(figsize=(15,5))
N_blocchi, I, error = np.loadtxt(r"Data/integrale_sampling.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(N_blocchi,I,yerr=error,color='limegreen')
plt.xlabel('N blocchi')
plt.ylabel('Integrale')
plt.title("I con Importance Sampling")
plt.axhline(y=1,color='forestgreen')
plt.grid(True)
plt.show()
```

### Esercizio 02.2
- calcolare $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ con $r$ pari alla distanza dall'origine al variare del numero di passi in un *Random Walk* a passo discreto e continuo

In entrambi i casi è stato utilizzato il metodo del Data Blocking: il walker esegue $100$ passi e per ciascun passo sono state eseguite $M=10^4$ simulazioni (ripartendo sempre dall'origine) divise in $N=100$ blocchi.
In ciascun blocco è stata calcolata la distanza media dall'origine e la media totale con errore viene aggiornata allo scorrere dei blocchi.

Per entrambi i RW la distanza media ha un andamento del tipo: $r = K \sqrt{n_{step}}$

### RW lungo un reticolo 3D

Il passo reticolare è $a=1$ e ogni passo può avvenire in avanti o in indietro, lungo le direzioni x, y e z.

*Nota sul codice:* per assegnare la direzione e il verso del passo è stato estratto un numero casuale con distribuzione uniforme $x \in [0,1]$, l'intervallo di estrazione è poi stato diviso in 6 intervalli e in base all'appartenenza di $x$ ad uno di essi è stata assegnata la direzione eil verso del passo.
In particolare, indicando con $i$ l'indice dell'intervallo compreso tra 1 e 6, in caso di $i \, pari$ il passo è stato compiuto in avanti e viceversa per $i \, dispari$.
Infine la direzione $x$ è assegnata a $i \in \{1,2\}$, $y$ a $i \in \{3,4\}$ e $z$ a $i \in \{5,6\}$ 

```python
def f(x,K):
    return K*np.sqrt(x)

fig=plt.figure(figsize=(15,10))

N_passi, r, error = np.loadtxt(r"Data/random_discreto.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(N_passi,r,yerr=error, color='forestgreen', label='Random Walk discreto')
plt.xlabel('N passi')
plt.ylabel('distanza media dall\'origine')
plt.title('Random Walk lungo un reticolo: distanza media dall\'origine')

x = np.linspace(0, 100, 100,  endpoint=True)
p_opt, p_cov = curve_fit(f, x, r)
y_fit = f(x,p_opt[0])
plt.plot(x,y_fit, linewidth='2.5', color='cyan', label='Fit') # plotting fitted function
print("optimized parameter K:")
print(p_opt)
print("parameters uncertainty:")
print(np.sqrt(np.diagonal(p_cov)))

plt.legend()
plt.grid(True)
plt.show()
```

### RW continuo in 3D

Ogni passo dista $r=1$ dal precedente.
Mediante l'utilizzo di coordinate sferiche ad ogni passo vengono estratti uniformemente $\phi \in [0,2\pi]$ e $T \in [0,1]$, con $\theta = arccos(1 - 2T)$, e alle coordinate cartesiane della posizione precedente vengono sommati i valori:

$$\begin{cases} dx = sen(\theta)cos(\phi) \\ dy = sen(\theta)sen(\phi) \\ dz = cos(\theta) \end{cases}$$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig=plt.figure(figsize=(15,10))

N_passi, r, error = np.loadtxt(r"Data/random_continuo.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(N_passi,r,yerr=error,color='forestgreen', label='Random Walk continuo')
plt.xlabel('N passi')
plt.ylabel('distanza media dall\'origine')
plt.title('Random Walk continuo: distanza media dall\'origine')

x = np.linspace(0, 100, 100,  endpoint=True)
p_opt, p_cov = curve_fit(f, x, r)
y_fit = f(x,p_opt[0])
plt.plot(x,y_fit,  linewidth='2.5', color='cyan', label='Fit') # plotting fitted function
print("optimized parameter K:")
print(p_opt)
print("parameters uncertainty:")
print(np.sqrt(np.diagonal(p_cov)))

plt.grid(True)
plt.show()
```
