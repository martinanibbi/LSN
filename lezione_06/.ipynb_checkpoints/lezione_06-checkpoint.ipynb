### Il modello di Ising

L'obiettivo di questa esercitazione è lo studio del modello di Ising $1D$ per un sistema di $50$ spin mediante il calcolo dell'energia interna, della capacità termica, della magnetizzazione e della suscettività al variare della temperatura T.
In particolare, la temperatura varia in un intervallo $T \in [0.1, 3.1]$ in step da $\Delta T = 0.1$.
L'energia interna, la capacità termica e la suscettività sono state calcolate con un campo magnetico esterno nullo, mentre per il calcolo della magnetizzazione è stato imposto un campo magnetico $h=0.02$.

Inoltre, è possibile confrontare i risultati teorici con le simulazioni relative a due diversi algoritmi: Metropolis e Gibbs.

Ogni simulazione parte dalla configurazione finale della simulazione precedente e, siccome gli step di temperatura sono sufficientemente piccoli, possiamo assumere che le configurazioni siano tra loro simili e dunque evitare l'equilibrazione del sistema per gli step successivi al primo.

Di seguito vengono riportati i risultati per il *Gibbs sampling* e *Metropolis sampling*, confrontati con il valore teorico del modello di Ising.
Ciascuna simulazione è stata eseguita con il metodo del *data blocking* mediante $20$ blocchi da $10000$ passi ciascuno. 

```python
import matplotlib                                        #definisco funzioni per dopo...
import matplotlib.pyplot as plt
import numpy as np

points = 100
J = 1.0
Ns = 50

def b(t):                                                #beta!
    return 1./t

def th(t):
    return np.tanh(J*b(t))

def thN(t):
    return th(t)**Ns
    
def ch(t):
    return 1./th(t)

def e(t):                                                 #ENERGIA INTERNA
    return -J*( th(t) + ch(t)*thN(t) )/( 1 + thN(t) )

def heat(t):                                              #CALORE SPECIFICO
    return ((b(t)*J)**2)*(((1+thN(t)+(Ns-1)*(th(t)**2)+(Ns-1)*(ch(t)**2)*thN(t))/(1+thN(t)))-Ns*((th(t)+ch(t)*thN(t))/(1+thN(t)))**2)

def l1(t,h):
    return np.exp(b(t)*J)*np.cosh(b(t)*h)+np.sqrt(np.exp(2*b(t)*J)*np.cosh(b(t)*h)*np.cosh(b(t)*h)-2*np.sinh(2*b(t)*J))

def l2(t,h):
    return np.exp(b(t)*J)*np.cosh(b(t)*h)-np.sqrt(np.exp(2*b(t)*J)*np.cosh(b(t)*h)*np.cosh(b(t)*h)-2*np.sinh(2*b(t)*J))

def Z(t,h):                                              
    return l1(t,h)**Ns + l2(t,h)**Ns
    
def M(t,h):                                               #MAGNETIZZAZIONE
    return (np.exp(b(t)*J)*np.sinh(b(t)*h)*((l1(t,h)**(Ns-1))*(1+np.exp(b(t)*J)*np.cosh(b(t)*h)/np.sqrt(np.exp(2*b(t)*J)*np.cosh(b(t)*h)*np.cosh(b(t)*h)-2*np.sinh(2*b(t)*J))) + (l2(t,h)**(Ns-1))*(1-np.exp(b(t)*J)*np.cosh(b(t)*h)/np.sqrt(np.exp(2*b(t)*J)*np.cosh(b(t)*h)*np.cosh(b(t)*h)-2*np.sinh(2*b(t)*J)))))/(Z(t,h))
def X(t):                                                 #SUSCIETTIVITA' (per h=0!!)
    return b(t)*np.exp(2*b(t)*J)*(1-thN(t))/(1+thN(t))
```

### Energia Interna  $U(N, T, h=0)$

```python
fig=plt.figure(figsize=(15, 7))                     #sx: larghezza; dx: Lunghezza
plt.title('Ising 1D, energia interna - h = 0')

T = np.linspace(0.1,3.1,num=points)
plt.plot(T, e(T), linewidth='3.5', color='violet', label="Theorical")
plt.xlabel('T')
plt.ylabel('U(T)')

temp, energy, error = np.loadtxt("Data/Metropolis/energy/final.dat", usecols=(0,1,2), unpack='true')
plt.errorbar(temp, energy, yerr=error, color ='purple', label="Metropolis")

temp, energy, error = np.loadtxt("Data/Gibbs/energy/final.dat", usecols=(0,1,2), unpack='true')
plt.errorbar(temp, energy, yerr=error, color ='red', label="Gibbs")

plt.grid(True)
plt.legend()    
plt.show()
```

### Calore Specifico  $C(N, T, h=0)$

```python
fig=plt.figure(figsize=(15, 7))                     #sx: larghezza; dx: Lunghezza
plt.title('Ising 1D, calore specifico - h = 0')

T = np.linspace(0.1,3.1,num=points)
plt.plot(T, heat(T), linewidth='3.5', color='violet', label="Theorical")
plt.xlabel('T')
plt.ylabel('C(T)')

temp, h_cap, error = np.loadtxt("Data/Metropolis/heat_capacity/final.dat", usecols=(0,1,2), unpack='true')
plt.errorbar(temp, h_cap, yerr=error, color ='purple', label="Metropolis")

temp, h_cap, error = np.loadtxt("Data/Gibbs/heat_capacity/final.dat", usecols=(0,1,2), unpack='true')
plt.errorbar(temp, h_cap, yerr=error, color ='red', label="Gibbs")

plt.grid(True)
plt.legend()    
plt.show()
```

### Magnetizzazione  $M(N,T, h=0.02)$

```python
fig=plt.figure(figsize=(15, 7))                     #sx: larghezza; dx: Lunghezza
plt.title('Ising 1D, magnetizzazione - h = 0.02')

T = np.linspace(0.1,3.1,num=points)
plt.plot(T, M(T,0.02), linewidth='3.5', color='violet', label="Theorical")
plt.xlabel('T')
plt.ylabel('M(T)')

temp, mag, error = np.loadtxt("Data/Metropolis/magnetization/final.dat", usecols=(0,1,2), unpack='true')
plt.errorbar(temp, mag, yerr=error, color ='purple', label="Metropolis")

temp, mag, error = np.loadtxt("Data/Gibbs/magnetization/final.dat", usecols=(0,1,2), unpack='true')
plt.errorbar(temp, mag, yerr=error, color ='red', label="Gibbs")

plt.grid(True)
plt.legend()   
plt.show()
```

### Suscettività    $\chi(N,T, h=0)$

```python
fig=plt.figure(figsize=(15, 7))                     #sx: larghezza; dx: Lunghezza
plt.title('Ising 1D, suscettività - h=0')

T = np.linspace(0.1,3.1,num=points)
plt.plot(T, X(T), linewidth='3.5', color='violet', label="Theorical")
plt.xlabel('T')
plt.ylabel('$\chi(T)$')

temp, sus, error = np.loadtxt("Data/Metropolis/susceptibility/final.dat", usecols=(0,1,2), unpack='true')
plt.errorbar(temp, sus, yerr=error, color ='purple', label="Metropolis")

temp, sus, error = np.loadtxt("Data/Gibbs/susceptibility/final.dat", usecols=(0,1,2), unpack='true')
plt.errorbar(temp, sus, yerr=error, color ='red', label="Gibbs")

plt.grid(True)
plt.legend()    
plt.show()
```
