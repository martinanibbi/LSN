```python
import matplotlib                                        #definisco funzioni per dopo...
import matplotlib.pyplot as plt
import numpy as np

temp = [0.5, 1.0, 1.5, 2.0]                              #(divido T in 4 intervalli)
vett_Metro = [0, 0, 0, 0]                                #salvo quantità all'ultimo blocco x grafici in funzione di T
vett_Gibbs = [0, 0, 0, 0]
err_Metro = [0, 0, 0, 0]
err_Gibbs = [0, 0, 0, 0]

points = 100
J = 1.0
Ns = 50

def b(t):                                                #beta!
    return 1./t

def th(t):
    return np.tanh(J*b(t))

def thN(t):
    return th(t)**Ns
    
def ch(t):
    return 1./th(t)

def e(t):                                                 #ENERGIA INTERNA
    return -J*( th(t) + ch(t)*thN(t) )/( 1 + thN(t) )

def heat(t):                                              #CALORE SPECIFICO
    return ((b(t)*J)**2)*(((1+thN(t)+(Ns-1)*(th(t)**2)+(Ns-1)*(ch(t)**2)*thN(t))/(1+thN(t)))-Ns*((th(t)+ch(t)*thN(t))/(1+thN(t)))**2)

def l1(t,h):
    return np.exp(b(t)*J)*np.cosh(b(t)*h)+np.sqrt(np.exp(2*b(t)*J)*np.cosh(b(t)*h)*np.cosh(b(t)*h)-2*np.sinh(2*b(t)*J))

def l2(t,h):
    return np.exp(b(t)*J)*np.cosh(b(t)*h)-np.sqrt(np.exp(2*b(t)*J)*np.cosh(b(t)*h)*np.cosh(b(t)*h)-2*np.sinh(2*b(t)*J))
    
def Z(t,h):                                              
    return l1(t,h)**Ns + l2(t,h)**Ns
    
def M(t,h):                                               #MAGNETIZZAZIONE
    return (np.exp(b(t)*J)*np.sinh(b(t)*h)*((l1(t,h)**(Ns-1))*(1+np.exp(b(t)*J)*np.cosh(b(t)*h)/np.sqrt(np.exp(2*b(t)*J)*np.cosh(b(t)*h)*np.cosh(b(t)*h)-2*np.sinh(2*b(t)*J))) + (l2(b(t),h)**(Ns-1))*(1-np.exp(b(t)*J)*np.cosh(b(t)*h)/np.sqrt(np.exp(2*b(t)*J)*np.cosh(b(t)*h)*np.cosh(b(t)*h)-2*np.sinh(2*b(t)*J)))))/(Z(t,h))

def X(t):                                                 #SUSCIETTIVITA' (per h=0!!)
    return b(t)*np.exp(2*b(t)*J)*(1-thN(t))/(1+thN(t))
```
