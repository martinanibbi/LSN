### Fit lineare

Consideriamo la retta:
$$
f(x)=2x+1
$$

con $x \in [-1,1]$ e con i parametri fissati:

```python
m = 2 # slope
b = 1 # intersect
```

L'obiettivo è trovare i parametri migliori con i quali eseguire il fit di $f(x)$, in particolare per questo esercizio vengono cambiati:
- il numero di epoche $ep$
- la deviazione standard $sigma$ del rumore Gaussiano aggiunto ai dati di test e training
- il numero di punti $n\_data$ con cui avviene il training della rete neurale. Il test viene sempre eseguito con un decimo di $n\_data$
- il batch size, indicato come $b\_size$, cioè la dimensione dei sottointervalli nel quale l'intero dataset di training viene diviso 

Di seguito viene eseguita una simulazione con i parametri fissati:
- $ep = 10$
- $sigma = 0.5$
- $n\_data = 500$
- $b\_size = 5$

```python
ep = 10
sigma = 0.5
n_data = 500
b_size = 5
```

```python
import numpy as np

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, n_data)
x_valid = np.random.uniform(-1, 1, int(n_data/10))
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function

y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(m * x_valid + b, sigma)
```

```python
# plot training and target dataset
import matplotlib.pyplot as plt
plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
plt.grid(True); 
plt.show()
```

Alla prima iterazione viene salvato il modello iniziale nella cartella *Saved* in modo tale da essere utilizzato anche durante le iterazioni successive.
Senza questo accorgimento verrebbe inizializzato un modello diverso ad ogni *Restart* del Kernel e ciò renderebbe più difficile osservare delle differenze variando i parametri iniziali.

Sebbene siano state testate strutture con più di un layer e più di un neurone, il risultati migliori sembrano coincidere proprio con la scelta di un unico layer dotato di un unico neurone.
Questo poichè l'azione di un singolo neurone è "lineare", e dunque una rete costituita da un unico strato con un unico neurone si adatta perfettamente al nostro problema.
Inoltre, testando i modelli con vari layer, si nota che tutte le funzioni di attivazione risultano inadeguate al problema.

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras
import os

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

save_model_path = 'Saved/line_model'

line_model = tf.keras.Sequential()
line_model.add(Dense(1, input_shape=(1,)))
line_model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

if os.path.isfile(save_model_path):
    line_model = tf.keras.models.load_model(filepath=save_model_path)

    for layer in line_model.layers:
            print(layer.get_weights())
            
line_model.save(filepath=save_model_path, include_optimizer=True)
line_model.summary()
```

```python
history = line_model.fit(x=x_train, y=y_train, 
          batch_size=b_size, epochs=ep,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
# return weights
line_model.get_weights()
```

```python
# evaluate model
score = line_model.evaluate(x_valid, y_valid, batch_size=b_size, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test Mean Square Error:', score[1])
```

```python
# evaluate model with the exact curve
score = line_model.evaluate(x_valid, y_target, batch_size=b_size, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test Mean Square Error:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
fig=plt.figure(figsize=(10,7))

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')

plt.grid(True)
plt.show()
```

```python
fig=plt.figure(figsize=(15,10))
x_predicted = np.random.uniform(-5, 5, 100)
y_predicted = line_model.predict(x_predicted)
#print(y_predicted.size)
plt.scatter(x_predicted, y_predicted,color='r')
xplot = np.linspace(-5,5,100)
plt.plot(xplot, m*xplot+b)
plt.grid(True)
plt.show()
```

```python
save_final_model_path='Saved/line_model_final'
line_model.save(filepath=save_final_model_path, include_optimizer=True)
```

### Considerazioni finali

La simulazione è stata eseguita con le combinazioni tra i seguenti parametri:
- $sigma = 0.1, 0.5, 1 $
- $n\_data = 100, 500, 1000 $
- $b\_size = 5, 10, 20$

Innanzitutto, per qualunque scelta di $sigma$ e $n\_data$ il risultato migliora sensibilmente al decrescere di $b\_size$.
Fissiamo dunque $b\_size = 5$ per l'ottimizzazione degli altri parametri.

Inoltre, $ep$ viene sempre scelto in modo tale da evitare l'overfitting dei dati e dunque il training viene fermato appena la funzione costo si stabilizza.

Di seguito i risultati per le combinazioni di $sigma$ e $n\_data$:
- $sigma = 0.1$: risultati buoni per $n\_data = 100$, ottimi per $n\_data = 500, 1000$
- $sigma = 0.5, 1$: risultati pessimi per $n\_data = 100$, buoni per $n\_data = 500$, ottimi per $n\_data = 1000$

Il modello rimane predittivo anche fuori dal range di test e di training per la natura stessa del problema.
I dati fuori dal range di training (come si può vedere facilmente nell'esercizio seguente) proseguono seguendo la direzione degli ultimi dati prodotti nell'intervallo ottimale.
Essendo la funzione da fittare una retta, il range di validità del modello si allarga in base alla bontà stessa del modello.

*Osservazione:* Il rumore Gaussiano aggiunto ai dati di training peggiora la performance della rete neurale e ciò può essere risolto aumentando il numero di dati disponibili (qualora sia possibile).
Tuttavia questo è più conforme ad una situazione reale, in quanto i dati sperimentali hanno sempre un rumore di fondo, e contribuisce a rendere il modello più generale.
In caso di dati di training perfettamente privi di errore, la rete converge velocemente ad un modello altamente specifico per per i dati utilizzati durante l'apprendimento e si verifica quasi subito un "overfitting" dei dati.
Con la presenza di errori sperimentali invece il modello converge più lentamente e ha una maggiore capacità di generalizzazione.

L'errore sperimentale risulta dunque vantaggioso, supponendo di poter aumentare i dati a proprio piacimento.

### Fit polinomiale

Consideriamo ora un polinomio di ordine 3:
$$
f(x)=4-3x-2x^2+3x^3
$$
con $x \in [-1,1]$.

La simulazione viene eseguita variando:

- il numero di layers
- il numero di neuroni in ogni layer
- la funzione di attivazione
- l'ottimizzatore
- la funzione *Loss*

```python
def f(x):
    return 4 - 3*x - 2*x**2 + 3*x**3
```

```python
ep = 20
sigma = 0.2
n_data = 1000
b_size = 32
```

```python
# generate training inputs
np.random.seed(2)
x_train = np.random.uniform(-1, 1, n_data)
x_valid = np.random.uniform(-1, 1, int(n_data/10))
x_valid.sort()
y_target = f(x_valid) # ideal (target) polynomial function


y_train = np.random.normal(f(x_train), sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(f(x_valid), sigma)
```

```python
# plot training and target dataset
import matplotlib.pyplot as plt
plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
plt.grid(True); plt.show()
```

```python
save_model_path = 'Saved/polynomial_model'

pol_model = tf.keras.Sequential()
pol_model.add(Dense(5, input_shape=(1,)))
pol_model.add(Dense(10, activation='elu'))
pol_model.add(Dense(10, activation='elu'))
pol_model.add(Dense(10, activation='elu'))
pol_model.add(Dense(10, activation='elu'))
pol_model.add(Dense(10, activation='elu'))
pol_model.add(Dense(10, activation='elu'))
pol_model.add(Dense(10, activation='elu'))
pol_model.add(Dense(10, activation='elu'))
pol_model.add(Dense(10, activation='elu'))
pol_model.add(Dense(10, activation='elu'))
pol_model.add(Dense(10, activation='elu'))
pol_model.add(Dense(10, activation='elu'))
pol_model.add(Dense(10, activation='elu'))
pol_model.add(Dense(5, activation='elu'))
pol_model.add(Dense(1, activation='elu'))
           
pol_model.compile(optimizer='Adam', loss='mae', metrics=['mae'])

if os.path.isfile(save_model_path):
    pol_model = tf.keras.models.load_model(filepath=save_model_path)

    for layer in pol_model.layers:
            print(layer.get_weights())
            
pol_model.save(filepath=save_model_path, include_optimizer=True)
pol_model.summary()
```

```python
history = pol_model.fit(x=x_train, y=y_train, 
          batch_size=b_size, epochs=ep,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
# return weights
pol_model.get_weights()
```

```python
# evaluate model
score = pol_model.evaluate(x_valid, y_valid, batch_size=b_size, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test Mean Square Error:', score[1])
```

```python
# evaluate model with the exact curve
score = pol_model.evaluate(x_valid, y_target, batch_size=b_size, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test Mean Square Error:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
fig=plt.figure(figsize=(10,7))

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.grid(True)
plt.show()
```

```python
fig=plt.figure(figsize=(15,10))

x_predicted = np.random.uniform(-1.5, 1.5, 200)
y_predicted = pol_model.predict(x_predicted)
#print(y_predicted.size)
plt.scatter(x_predicted, y_predicted,color='r')
xplot = np.linspace(-1.5,1.5,200)
plt.plot(xplot, f(xplot))
plt.grid(True)
plt.show()
```

```python
save_final_model='Saved/final_polynomial_model'
pol_model.save(filepath=save_final_model, include_optimizer=True)
```

### Considerazioni finali

La capacità rappresentativa della rete neurale aumenta sia con il numero di neuroni per strato sia con la profondità della rete stessa.
I risultati risultano simili nei casi:
- neuroni: ~25; strati: ~4
- neuroni: ~10; strati: ~15
- neuroni: ~25; strati: ~15

È inoltre possibile notare che la performance della rete migliora se il numero di neuroni è simile in ogni strato, mentre creare un picco di neuroni al centro della rete risulta meno performante.

La *elu* risulta essere la migliore funzione di attivazione, la *selu* funziona quasi alltrettanto bene mentre la rete fallisce totalmente con l'utilizzo delle funzioni *sigmoid*, *tanh*, *softsign* e *relu*.

Si ottengono inoltre risultati migliori se al primo layer non viene fornita alcuna funzione di attivazione.

Sono stati testati gli ottimizzatori Adadelta, Adagrad, Sgd, Adam e non risultano sostanziali differenze, mentre Rmsprop sembra non funzionare per questo problema.

Infine, la scelta migliore per la funzione Loss (e per la metrica) sembra essere la *Mean Absolute Error*, anche se utilizzando la *Mean Square Error* e *Mean Absolute Percentage Error* non si ottengono sostanziali differenze, mentre il *Mean Squared Logarithmic Error* non funziona per questo problema.

Il modello non risulta predittivo al di fuori del range di training e la rete fatica a risolvere il minimo vicino all'estremo destro del range di apprendimento.

### Funzione trigonometrica 2D
  
Infine, eseguiamo il fit della seguente funzione trigonometrica 2D: $$f(x,y) = \sin(x^2+y^2)$$ 
nell'intervallo $x \in [-3/2,3/2]$ e $y \in [-3/2,3/2]$.

```python
def f(x,y):
    return np.sin(x*x + y*y)
```

```python
ep = 200
sigma = 0.2
n_data = 1000
b_size = 5
```

```python
# generate training inputs
np.random.seed(1)
x_train = np.random.uniform(-1.5, 1.5, n_data)
x_valid = np.random.uniform(-1.5, 1.5, int(n_data/10))
y_train = np.random.uniform(-1.5, 1.5, n_data)
y_valid = np.random.uniform(-1.5, 1.5, int(n_data/10))
z_target = f(x_valid, y_valid) # ideal (target) polynomial function

z_train = np.random.normal(f(x_train,y_train), sigma) # actual measures from which we want to guess regression parameters
z_valid = np.random.normal(f(x_valid,y_valid), sigma)
```

```python
# plot training and target dataset
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure(figsize=(15,10))
ax = Axes3D(fig)

X = np.linspace(-1.5, 1.5, n_data)
Y = np.linspace(-1.5, 1.5, n_data)
X, Y = np.meshgrid(X, Y)
Z = f(X,Y)

ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False, alpha=0.5)
ax.scatter(x_valid, y_valid, z_valid, color='r')
#ax.view_init(10, 30)
ax.grid(True)
plt.show()
```

```python
# compose the NN model

save_model_path = 'Saved/sin_model'

sin_model = tf.keras.Sequential()
sin_model.add(Dense(3, input_shape=(2,)))
sin_model.add(Dense(10, activation='elu' ))
sin_model.add(Dense(20, activation='elu' ))
sin_model.add(Dense(20, activation='elu' ))
sin_model.add(Dense(20, activation='elu' ))
sin_model.add(Dense(20, activation='elu' ))
sin_model.add(Dense(20, activation='elu' ))
sin_model.add(Dense(20, activation='elu' ))
sin_model.add(Dense(20, activation='elu' ))
sin_model.add(Dense(20, activation='elu' ))
sin_model.add(Dense(20, activation='elu' ))
sin_model.add(Dense(20, activation='elu' ))
sin_model.add(Dense(7, activation='elu'))
sin_model.add(Dense(1, activation='elu'))
           
sin_model.compile(optimizer='adadelta', loss='mae', metrics=['mae'])

if os.path.isfile(save_model_path):
    sin_model = tf.keras.models.load_model(filepath=save_model_path)

    for layer in sin_model.layers:
            print(layer.get_weights())
            
sin_model.save(filepath=save_model_path, include_optimizer=True)
sin_model.summary()
```

```python
data_train = np.transpose([x_train, y_train])
data_valid = np.transpose([x_valid, y_valid])

history = sin_model.fit(x=data_train, y=z_train, 
          batch_size=b_size, epochs=ep,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(data_valid, z_valid))
```

```python
score = sin_model.evaluate(data_valid, z_valid, batch_size=b_size, verbose=1)
# print performance
print()
print('Test loss:', score[0])
print('Test Mean Square Error:', score[1])
```

```python
# evaluate model
score = sin_model.evaluate(data_valid, z_target, batch_size=b_size, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test Mean Square Error:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
fig = plt.figure(figsize=(15,10))
ax = Axes3D(fig)

fig=plt.figure(figsize=(15,10))

x_predicted = np.random.uniform(-1.5, 1.5, 1000)
y_predicted = np.random.uniform(-1.5, 1.5, 1000)
data_predicted = np.transpose([x_predicted, y_predicted])
z_predicted = sin_model.predict(data_predicted)
ax.scatter(x_predicted, y_predicted, z_predicted, color='darkblue')
ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False, alpha=0.5)
#ax.view_init(10, 10)
plt.show()
```
