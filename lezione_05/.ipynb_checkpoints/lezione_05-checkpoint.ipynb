Con l'algoritmo di Metropolis è possibile campionare una qualunqe distribuzione di probabilità.
Ciò risulta particolarmente utile in meccanica quantistica poichè il modulo quadro di una funzione d'onda è una densità di probabilità.

Un esempio è dato dalle autofunzioni dell'atomo di Idrogeno (in coordinate sferiche):

$$
\Psi_{n,l,m}(r,\theta,\phi)=
\sqrt{\left(\frac{2}{na_0}\right)^3 \frac{(n-l-1)!}{2n[(n+l)!]}}
e^{-r/na_0}\left(\frac{2r}{na_0}\right)^l L_{n-l-1}^{2l+1}\left(\frac{2r}{na_0}\right)
Y_l^m(\theta,\phi)
$$

In particolare sono riportati di seguito le posizioni campionate in 3D di un elettrone in un atomo di idrogeno negli stati 1s, 2p, 3d, 4f.
Nell'algoritmo di Metropolis sono state utilizzate una distribuzione uniforme e Gaussiana, i cui risultati sono equivalenti.

**Nota: ** l'equilibrazione è stata eseguita con $N = 10^2$ punti, mentre tutte le simulazioni sono state eseguite con $M = 10^6$ punti divisi in 100 blocchi

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

### Equilibrazione

Prima di procedere con la simulazione dobbiamo verificare che la posizione iniziale e la lunghezza del passo $\delta$ garantiscano un'accettazione media dell'algoritmo di Metropolis vicino al 50%.
La posizione iniziale, ragionevole per entrambi gli stati 1s e 2p, è: $(0.5, 0.7, 1.2)$.

*Nota:* L'equilibrazione è stata eseguita utilizzando la distribuzione uniforme, ma risulta ottimale anche per la distribuzione Gaussiana.

```python
fig=plt.figure(figsize=(16, 5))                     #sx: larghezza; dx: Lunghezza
fig.subplots_adjust(hspace=0.4, wspace=0.4)          #spazio tra due grafici vicini


data_input = open("Data/input.dat", 'r')   #primi 3 elementi sono le coordinate iniziali
start_pos = data_input.readlines()
    
print("Posizione iniziale per gli stati 1s e 2p:")
print("x: " + start_pos[0])
print("y: " + start_pos[1])
print("z: " + start_pos[2])

plt.subplot(1,2,1)
delta, accettazione = np.loadtxt("Data/1s/Uniform/equilibration.dat", usecols=(1,2), delimiter=' ', unpack='true')
plt.scatter(delta, accettazione, color='crimson')
plt.axhline(y=0.5,color='lime')
plt.xlabel('delta')
plt.ylabel('Accettazione')
plt.title('Accettazione in fase di equilibrazione per lo stato 1s - distribuzione Uniforme')
plt.grid(True)

plt.subplot(1,2,2)
delta, accettazione = np.loadtxt("Data/2p/Uniform/equilibration.dat", usecols=(1,2), delimiter=' ', unpack='true')
plt.scatter(delta, accettazione, color='crimson')
plt.axhline(y=0.5,color='lime')
plt.xlabel('delta')
plt.ylabel('Accettazione')
plt.title('Accettazione in fase di equilibrazione per lo stato 2p - distribuzione Uniforme')
plt.grid(True)

plt.show()
```

### Stato 1s

**$\delta$ per lo stato 1s: ** 1.5

```python
fig=plt.figure(figsize=(16, 10))                     #sx: larghezza; dx: Lunghezza
fig.subplots_adjust(hspace=0.2, wspace=0.4)          #spazio tra due grafici vicini

plt.subplot(2,1,1)
N_blocchi, r_1s_u, error = np.loadtxt(r"Data/1s/Uniform/radius.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(N_blocchi, r_1s_u, yerr=error)
plt.axhline(y=1.5,color='crimson',linewidth='2.5')
plt.xlabel('N blocchi')
plt.ylabel("<r>")
plt.title('Distanza media dall\'origine nello stato 1s (distribuzione uniforme)')
plt.grid(True)

plt.subplot(2,1,2)
N_blocchi, r_1s_g, error = np.loadtxt(r"Data/1s/Gauss/radius.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(N_blocchi, r_1s_g, yerr=error)
plt.axhline(y=1.5,color='crimson',linewidth='2.5')
plt.xlabel('N blocchi')
plt.ylabel("<r>")
plt.title('Distanza media dall\'origine nello stato 1s (distribuzione Gaussiana)')
plt.grid(True)

plt.show()
```

```python
fig = plt.figure(figsize=(10,10))

X, Y, Z =  np.loadtxt("Data/1s/Uniform/position.dat", usecols=(0,1,2), delimiter=' ', unpack='true')

ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.title('Campionamento della posizione nello stato 1s (distribuzione Uniforme)')

plt.show()
```
