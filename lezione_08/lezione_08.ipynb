L'obiettivo di questa esercitazione è trovare il Ground State per una particella unidimensionale del seguente potenziale esterno: 
$$
V(x) = x^4 - \frac{5}{2}x^2
$$
Il metodo utilizzato è il *Variational Monte Carlo*.
Innanzitutto, viene scelta una funzione d'onda di prova parametrizzata, nel nostro caso: 

$$ \Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}} + e^{-\frac{(x+\mu)^2}{2\sigma^2}}$$

A $\mu$ e $\sigma$ fissati viene campionata la densità di probabilità pari a  $ |\Psi_T^{\sigma,\mu}(x)|^2 $ attraverso un algoritmo di Metropolis ed essa viene utilizzata per calcolare il seguente integrale: 

$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} = 
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}
$$

Il valore dell'energia così trovato è sicuramente maggiore rispetto all'energia di Ground State, dunque vengono variati i parametri $\mu$ e $\sigma$ in modo tale da minimizzare il valore dell'integrale e con i suddetti valori viene infine campionata $ |\Psi_T^{\sigma,\mu}(x)|^2 $, la quale corrsiponde al modulo quadro della funzione d'onda più vicina possibile al Ground State.

### Minimizzazione dell'energia attraverso la variazione dei parametri $\mu$ e $\sigma$

Il valore di $\langle {\hat H} \rangle_T$ viene calcolato attraverso il metodo del data blocking utilizzando $N=100$ blocchi, ciascuno composto da $M=10^3$ punti.
Per ogni simulazione, che viene eseguita con i parametri $\mu$ e $\sigma$ fissati, viene salvato il valore all'ultimo blocco di $\langle {\hat H} \rangle_T$.

Viene così creata una griglia 21x21 dove viene selezionata la coppia migliore di parametri $(\mu,\sigma)$.
Di seguito sono riportati i valori di $\langle {\hat H} \rangle_T$ al variare di $(\mu,\sigma)$ e il calcolo di $\langle {\hat H} \rangle_T$ al variare del numero di blocchi con la miglior coppia di parametri precedentemente trovata.

Possiamo ora confrontare il campionamento di $ |\Psi_T^{\sigma,\mu}(x)|^2 $ ottenuto con i migliori parametri $(\mu,\sigma)$ con l'autofunzione del Ground State esatto.

```python
def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 6
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)
```

```python
# Plot a few things
plt.figure(figsize=(15,5))

print("Ground state energy: ", E[0])
print("1st excited state energy: ", E[1])
print("2nd excited state energy: ", E[2])

scale = 0.3
plt.plot(x, scale*V, color="Black", label="Potential") # plot the potential
plt.plot(x,(psi[0])**2)
plt.plot(x,(psi[1])**2)
plt.plot(x,(psi[2])**2)
plt.title("Potential & Probabilities")
plt.xlabel("x")
plt.grid(True)
plt.xlim((-3,3))
plt.ylim((-0.6,0.6))
plt.show()
```

```python
fig=plt.figure(figsize=(15, 5))

def f(x,m,s,N):  # defining the fitting function
    e1 = np.exp(-(x-m)*(x-m)/(s*s))
    e2 = np.exp(-(x+m)*(x+m)/(s*s))
    amp = e1 + e2
    return N*amp*amp

nbins = 600
psi_x = np.loadtxt("Data/psi_sampling.dat", unpack = 'true')
plt.xlabel("x")
plt.title("Ground State")
plt.hist(psi_x, nbins, label='Variational Monte Carlo', color='darkblue')

#fit con funzione d'onda di prova (non conosco la 'normalizzazione'...)
bins = np.linspace(-3, 3, nbins)
data_entries, bins_1 = np.histogram(psi_x, bins=bins)
binscenters = np.array([0.5 * (bins[i] + bins[i+1]) for i in range(len(bins)-1)])  #vettore con posizioni == centro dei bin
popt, pcov = curve_fit(f, xdata=binscenters, ydata=data_entries) 
print(popt)
xspace = np.linspace(-3, 3, 1000)
#plt.bar(binscenters, data_entries, width=bins[1] - bins[0], color='navy', label=r'Histogram entries')
plt.plot(xspace, f(xspace, *popt), color='darkorange', linewidth=2.5, label=r'Fitted function')

#fit con soluzione esatta (non conosco la 'normalizzazione'...)
binlength = bins[2]-bins[1]
plt.plot(xspace, 100000*binlength*(psi[0])**2, label='exact Ground State', linewidth=2.5, color='violet')
      #moltiplico i dati della sol esatta per la lunghezza dei bin e il numero di dati usati nel campionamento (M=100000)


leg=plt.legend()
plt.grid(True)
plt.show()
```
