1) minimizzazione dell'energia attraverso la variazione dei parametri $\mu$ e $\sigma$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig=plt.figure(figsize=(16, 10))                     #sx: larghezza; dx: Lunghezza

mu, sigma, energy = np.loadtxt(r"Data/final_energy.dat", usecols=(0,1,2), unpack='true')

ax = Axes3D(fig)
ax.scatter(mu, sigma, energy, c=energy, marker='o')
ax.set_xlabel('$\mu$')
ax.set_ylabel('$\sigma$')
ax.set_zlabel('Eergia')
ax.view_init(30, 60)
plt.show()
```
