### European options

In un'opzione europea il *writer* garantisce all'*holder* il diritto di comprare (nel caso di una *call option*) un asset ad un tempo T e ad un prezzo fisso K detto *strike price*.
Alla scadenza, l'*holder* compra l'azione se il prezzo reale dell'asset supera il prezzo pattutito in precedenza e dunque il profitto finale può essere definito come:
$$profit = max[0, S(T) - K]$$

Il prezzo dell'opzione deve essere inoltre scontato di $e^{-rT}$ e dunque il prezzo dell'opzione risulta essere:

$$ \langle e^{-rT} (S(T) - K) \rangle$$

dove il valor medio può essere calcolato campionando $S(T)$ con un Monte Carlo e il metodo del data blocking.
Nel caso di una *put option* valgono le stesse considerazioni, con un profitto definito come $max[0, K - S(T)]$.

Di seguito vengono riportati i prezzi delle *call option* e *put option* calcolati campionando direttamente $S(t)$ al tempo finale e suddividendo l'intervallo $[0,T]$ in 100 tempi intermedi.
I parametri utilizzati sono i seguenti:
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$


### Call option, campionamento diretto

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig=plt.figure(figsize=(15,5))

N_blocchi, price, error = np.loadtxt(r"Data/Call_direct.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(N_blocchi,price,yerr=error, color='darkblue', label='Call option, direct')
plt.xlabel('N blocchi')
plt.ylabel('prezzo')
plt.axhline(y=14.975790778311286, color='red', label='Call option, Black-Scholes')

plt.legend()
plt.grid(True)
plt.show()
```
